```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data Set

```python
df = pd.read_excel("Supermarket Sales - Python Course.xlsx")
df
```

```python
df.info()
```

```python
df.columns
```

# Data Preparation

```python
#mengubah tipe data 
df_cleaned = df.copy()

df_cleaned["Date"] = pd.to_datetime(df_cleaned["Date"])
df_cleaned["Revenue"] = df_cleaned["Unit price"]*df_cleaned["Quantity"]

print(df_cleaned.info())
df_cleaned.head()
```

```python
# Check typo
df_cleaned["Gender"].value_counts()
```

```python
for col_name in["Date","Payment","Invoice ID","Branch","City","Customer type","Gender","Product line","Unit price","Quantity"]:
    print(df_cleaned[col_name].value_counts(),"\n")
```

```python
#replace
dict_typo = {"male":"Male",
            "female":"Female",
            "fml":"Female"}

df_replace = df_cleaned.replace(dict_typo)
df_replace.head(10)
```

```python
df_replace["Invoice ID"].duplicated().sum()
```

# EDA

By Date

```python
agg_dialy = df_replace.groupby("Date", as_index=False)["Revenue","Quantity"].sum()
agg_dialy.head()
```

```python
plt.rcParams["figure.figsize"]=(20,5)

plt.plot("Date","Revenue", data=agg_dialy)
plt.title('Revenue trend')
plt.xlabel("Date")
plt.ylabel("Rev($)")
plt.show()
```

```python
sns.lineplot(data=agg_dialy,x="Date",y="Quantity")
```

by payment

```python
agg_payment = df_replace.groupby("Payment", as_index=False).agg({"Revenue":"sum","Invoice ID":"nunique"})
agg_payment.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10, 5)

# Misalkan 'Payment' dan 'Revenue' adalah nama kolom dalam DataFrame agg_payment
sns.barplot(x="Payment", y="Revenue", data=agg_payment.sort_values("Revenue"))

plt.show()
```

dari count transaction ewalet paling banyak

```python
plt.pie(agg_payment["Revenue"],labels=agg_payment["Payment"],autopct='%1.1f%%')
```
